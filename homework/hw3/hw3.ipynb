```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import math 

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.datasets import fetch_openml
dataset = fetch_openml(name='boston', version=1, parser='auto')
print("Số chiều dữ liệu input: ", dataset.data.shape)
print("Số chiều dữ liệu target: ", dataset.target.shape)
print()

print("5 mẫu dữ liệu đầu tiên:")
print("input: ", dataset.data[:5])
print("target: ",dataset.target[:5])
```

```python
from sklearn.model_selection import train_test_split
X=dataset.data
y=dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state = 51)
```

```python
regr=linear_model.LinearRegression()
```

```python
def linear_regression(X_train,y_train):
    X_train = np.asarray(X_train, dtype=float)
    y_train = np.asarray(y_train, dtype=float)
    one = np.ones((X_train.shape[0],1))
    Xbar = np.concatenate((one, X_train), axis = 1)
    A = np.dot(Xbar.T, Xbar)
    b = np.dot(Xbar.T, y_train)
    w_lr = np.dot(np.linalg.pinv(A), b)
    coef = w_lr[1:]
    intercept = w_lr[0]
    return coef, intercept
```

```python
def predict(intercept, coef, X_test):
    return np.dot(X_test, coef) + intercept
```

```python
regr.fit(X_train,y_train)
print("[w1, ... w_n] = ", regr.coef_)
print("w0 = ", regr.intercept_)
```

```python
coef, intercept=linear_regression(X_train=X_train, y_train=y_train)
print ('coef=',coef)
print ('intercept=', intercept)
```

```python
y_pred = regr.predict(np.asarray(X_test, dtype=float))
print(y_pred)
pd.DataFrame(data=np.array([y_test, y_pred,abs(y_test-y_pred)]).T,columns=["Thực tế", "Dự đoán", "Lệch"])
```

```python
y_test_linear=predict(intercept=intercept, coef=coef, X_test=np.asarray(X_test, dtype=float))
print (y_test_linear)
pd.DataFrame(data=np.array([y_test, y_test_linear,abs(y_test-y_test_linear)]).T,columns=["Thực tế", "Dự đoán", "Lệch"])
```

```python
loss = math.sqrt(mean_squared_error(y_test, y_pred))
print("lỗi :",loss)
```

```python
loss = math.sqrt(mean_squared_error(y_test, y_test_linear))
print("lỗi :",loss)
```
